### ANN on PIMA dataset to predict onset of diabetes

```python
#Load required packages
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Set random seed for reproducibility
np.random.seed(2)
```

```python
# Load pima indians diabetes dataset
dataset = np.loadtxt("C:/Users/Sija/Downloads/pima.csv", delimiter=",")
```

```python
# Split into input (X) and output (Y) variables (Last row is the target)
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# Split X, Y into a train and test sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create model, add layers and specify activation function
model = Sequential()
model.add(Dense(15, input_dim=8, activation='relu')) # input_dim param is mandatory for input layer
model.add(Dense(10, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dropout(.2)) # to prevent over-fitting
model.add(Dense(1, activation='sigmoid')) # sigmoid function instead of relu for final layer 
```

```python
# Compile the model with loss and optimizer - adam gradient descent 
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
# Fit the model to the data (training the network)
model.fit(x_train, y_train, epochs = 1000, batch_size=20, validation_data=(x_test, y_test))
```

```python
#Evaluate model performance
scores = model.evaluate(X,Y)
print(model.metrics_names[1], scores[1]*100)
```

#### The model gives an accuracy of 81%
